```python
import pandas as pd
from jieba import analyse
tfidf = analyse.extract_tags
from pandas import DataFrame, Series
```

```python
#倒入file内文件
file = pd.read_excel('中心词标注语料.xlsx', header = 0)
```

```python
#创建DataFrame列表，并填充空值
df = pd.DataFrame({'Raw_data': file["Raw_data"],
                    'Results': file["Results"]} )

df['Results'] = df['Results'].fillna('我是空的！')
```

```python
#选一个值进行实验，并将值转换成String值
for n in range(0,len(df["Raw_data"])):
    c = df.loc[n, "Raw_data"]
    print(c)
```

```python
#通过tfidf算法，得出关键词然后转成String值
for n in range(0,len(df["Raw_data"])):
    c = df.loc[n, "Raw_data"]
    keywords = tfidf(c, topK = 5, withWeight = False, allowPOS =())
    final = "，".join(keywords)
    df.iat[n,1] = final
```
